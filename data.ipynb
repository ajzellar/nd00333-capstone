```python
import pandas as pd
import numpy as np

file_path = "salary.csv"
df = pd.read_csv(file_path)
df.info()
```

```python
for col in df:
  unique_vals = df[col].unique()
  print(f"{col} - {len(unique_vals)}")
  print(unique_vals)

# Remove ? from columns
df.replace(' ?', np.NaN, inplace=True)
df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
for col in df:
  df[col].fillna(value="Other", inplace=True)

if 'salary' in df.columns:
  y = df['salary'].map( {' >50K': 1, ' <=50K': 0} )
  df.drop('salary', axis=1, inplace=True)
if 'education' in df.columns:
  df.drop('education', axis=1, inplace=True)
num_cols = [x for x in df.columns if df[x].dtype != 'object']
ss = StandardScaler()
df[num_cols] = ss.fit_transform(df[num_cols])
df.info()
```

```python
import json
import os
obj_columns = list()
non_obj_columns = list()
for col in df.columns:
  if df[col].dtype == 'object':
    df[col] = df[col].str.strip()
    obj_columns.append(col)
  else:
    non_obj_columns.append(col)

df_out = pd.get_dummies(df[obj_columns])

output = dict()
for col in obj_columns:
  new_cols =  [nc.split("_")[-1] for nc in df_out.columns if col in nc]
  output[col] = new_cols

# Save mapping of old columns to new columns in order to help building requests
# with open('hot_encode_map.json', 'w') as f:
#   f.write(json.dumps(output, indent=2))

df_out[non_obj_columns] = df[non_obj_columns]
df_out
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_out , y , test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report


lr_model = LogisticRegression()
svm_model = SVC()
knn_model = KNeighborsClassifier()
models = [('lr', lr_model), ('svm', svm_model), ('knn', knn_model)]
for m_name, m in models:
  print(m_name)
  m.fit(X_train, y_train)
  y_preds = m.predict(X_test)
  print(m.score(X_test, y_test))
  print(roc_auc_score(y_test, y_preds))
  print(accuracy_score(y_test, y_preds))
  print(classification_report(y_test, y_preds))
```
